## 데이터 준비

```python
df = read_excel("https://data.hossam.kr/E04/diabetes.xlsx")
df.head()
```

## 독립변수만 추출

종속 변수 drop

주성분 분석의 정확도를 높이기 위해서는 표준화(데이터 정규화) 필수

독립변수만 뽑은 DF를 주로 x_train이라고 명명

```python
x_train = df.drop(['target'], axis=1)
x_train.head()
```

## 전체 요인에 대한 `회귀분석`

```python
xnames = list(df.columns)
xnames.remove('target')

model1 = my_ols(df, y='target', x=xnames)
model1
```

## 요약 결과 생성

```python
model1.summary
```

## 결과표 생성

```python
model1.table
```

## 예측값 생성

predict

```python
y1 = model1.fit.predict(x_train)
y1
```

## 차원 축소 적용

### `PCA 분석`

```python
# 종속변수 topfeat
pca_model = pca(n_components=len(x_train.columns))
fit = pca_model.fit_transform(x_train)
fit['topfeat']
```

### 시각화를 통한 기여도 확인

종속변수에 영향을 크게 미치는 요소를 찾기 위해` 보통 70~80% 지점까지의 변수`를 취함

여기서는 특정 변수의 value를 역순으로 정렬 후 시각화한 경우 급하강 직전 변수를 취한다

```python
pca_df = fit['topfeat']
pca_df.sort_values('loading', ascending=False, inplace=True)
pca_df.set_index('feature', inplace=True)
pca_df['loading']

xlabel = pca_df['loading'].index
y = pca_df['loading'].values
x = list(range(0, len(y)))
sb.lineplot(x=x, y=y)
plt.xticks(x, xlabel, rotation=0)
plt.show()
plt.close()
```

### 선정된 요인(변수)을 사용하여 회귀분석 수행

변수 확인, 추출은 수동(중복없이)

```python
model2 = my_ols(df, y='target', x=['s6','s4','s1','s5','bp','s3'])
```

### 요약 결과

```python
model2.summary
```

설명력(R-squared 값)이 다소 떨어짐

### 결과표

```python
y2 = model2.fit.predict(x_train)
y2
```

### 두 가지 모델의 결과 비교

```python
# DF의 변수, 변수 값 설정
rdf = DataFrame({
    's6':df['s6'],
    'y':df['target'],
    'y1':y1,
    'y2':y2
})

rdf
```

melt 함수를 사용해 DF 배치를 변경, 가독성을 높일 수 있음. (형식은 카테고리로 변경됨)

적용 전

| 이름 | 아침 | 점심 | 저녁 |
| -- | -- | -- | -- |
| A | 100m | 200m | 300m |

적용 후

| A  | 아침 | 100m |
| -- | -- | -- |
| A | 점심 | 200m |
| A | 저녁 | 300m |

id_vars=이름, value_vars=아침,점심,저녁

>참고 : D. 탐색적 데이터 분석/01. 데이터 시각화/09-서브플롯_예제(전국_지역별_실업률_변화).ipynb 

```python
mdf1 = melt(rdf, id_vars=['s6'], value_vars=['y', 'y1'])
mdf1
```

```python
mdf2 = melt(rdf, id_vars=['s6'], value_vars=['y', 'y2'])
mdf2
```

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.unicode_minus"] = False
```

```python
sb.lmplot(data=mdf1, x="s6", y="value", hue="variable", height=5, aspect=3)
plt.show()
plt.close()
```

```python
sb.lmplot(data=mdf2, x="s6", y="value", hue="variable", height=5, aspect=3)
plt.show()
plt.close()
```

# 문제1 (스케일링)

## 데이터 준비

```python
df = read_excel("https://data.hossam.kr/E04/diabetes.xlsx")
df.head()
```

## 데이터 표준화

머신러닝의 학습 시간과 성능의 개선을 위해 표준화 진행

### `독립변수`에 대한 표준화

```python
x_train = df.drop(['target'], axis=1)
x_train_std=StandardScaler().fit_transform(x_train)
x_train_std_df = DataFrame(x_train_std, columns=x_train.columns)
x_train_std_df.head()
```

### `종속변수`에 대한 표준화

```python
y_train = df.filter(['target'], axis=1)
y_train_std = StandardScaler().fit_transform(y_train)
y_train_std_df = DataFrame(y_train_std, columns=y_train.columns)
y_train_std_df
```

## `전체 요인`에 대한 `회귀분석`

```python
mdf = merge(x_train_std_df, y_train_std_df, left_index=True, right_index=True)

xnames = list(x_train_std_df.columns)
model1 = my_ols(mdf, y = 'target', x = xnames)
model1.summary
```

### 결과표

```python
model1.table
```

## `차원 축소` 적용

### `PCA 분석`

```python
pca_model = pca()
fit = pca_model.fit_transform(x_train_std_df)
topfit = fit['topfeat']
topfit
```

### 선정된 요인 이름 추출

```python
best = topfit.query("type=='best'")
feature = list(set(list(best['feature'])))
feature
```

### 선정된 요인을 사용하여 회귀분석 수행

```python
mdf = merge(x_train_std_df, y_train_std_df, left_index=True, right_index=True)
model2 = my_ols(mdf, y='target', x=feature)
model2.summary
```

## 두 가지 모델의 결과 비교

### 결과 비교를 위한 데이터프레임 생성

```python
# DF 설정
rdf = DataFrame({
    's6': mdf['s6'],
    'y': mdf['target'],
    'y1': model1.fit.predict(x_train_std_df),
    'y2': model2.fit.predict(x_train_std_df)
})

rdf
```

### melt 함수를 사용해 실제 값과 model1의 예측 결과를 재배치

```python
mdf1 = melt(rdf, id_vars=['s6'], value_vars=['y', 'y1'])
mdf1
```

### 실제 값과 model2의 예측 결과 재배치

```python
mdf2 = melt(rdf, id_vars=['s6'], value_vars=['y', 'y2'])
mdf2
```

### 각각의 재배치 결과를 시각화

```python
# lmplot은 통합적으로 hue와 추세선을 그릴 수 있는 그래프 함수
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["axes.unicode_minus"] = False

sb.lmplot(data=mdf1, x="s6", y="value", hue="variable", height=5, aspect=3)
sb.lmplot(data=mdf2, x="s6", y="value", hue="variable", height=5, aspect=3)

plt.show()
plt.close()
```
