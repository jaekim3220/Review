## #03. 한글 사용하기

Matplostib에서 pyplot객체에 컴퓨터에 깔려있는 한글 폰트를 설정해야 사용 가능

폰트를 다운받을 경우(윈도우는 TTF 형식으로 다운)

### 1. 컴퓨터에 설치되어 있는 글꼴 확인

```python
import os
import sys
from matplotlib import font_manager
```

```python
# 시스템의 글꼴 목록을 리스트로 추출
font_list = font_manager.findSystemFonts()

# 이릅순 정렬
font_list.sort()

# 각 폰트 파일의 경로에 대한 반복
for font_path in font_list:
    # 폰트 파일의 경로를 사용하여 '폰트 속성' 가져오기
    fp = font_manager.FontProperties(fname = font_path)

    try:
        font_name = fp.get_name()
        print(font_name)
    except:
        continue
```

### 2. 글꼴을 적용하기 위한 그래프 전역 설정

아래의 코드를 모든 그래프를 그리기 추가해 그래프 설정

```python
# 위에서 찾은 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 글자 크기
plt.rcParams['font.size'] = 12

# 그래프 크기
plt.rcParams['figure.figsize'] = (8,4)

# 유니코드 음수 부호 설정(그래프에 음수 추가 가능)
plt.rcParams['axes.unicode_minus'] = False
```

### 3. 한글, 음수 테스트 
font_manager, axes.unicode_minus

```python
plt.figure()    # 그래프 시작을 선언
plt.title("한글과 음수 테스트")    # 그래프 제목 표시
plt.plot([10,20,30,40], [1,2,3,4])   # 그래프 그리기
plt.xlabel("엑스 축 라벨") # x축 제목
plt.ylabel("와이 축 라벨") # y축 제목
plt.show()  # 그래프 화면 표시
plt.close() # 그래프 종료
```

### 4. 각 영역별로 개별 폰트 지정
xlabel, ylabel

```python
# windows font 설정(다르게 표시할 그래프의 숫자 만큼)
font1 = {'family': 'Batang', 'size': 24, 'color': '#2aa2c9', 'weight': 'bold'}
font2 = {'family': 'Gulim', 'size': 12, 'color': 'red'}
font3 = {'family': 'Gulim', 'size': 12, 'color':  '#ff6600', 'weight': 'light'}
```

```python
plt.figure()    # 그래프 시작을 선언
plt.title("서로 다른 글꼴 실습", fontdict=font1)    # 그래프 제목 표시
plt.plot([10,20,30,40,50],[1,4,10,7,15])   # 그래프 그리기
plt.xlabel("엑스 축", fontdict=font2) # x축 제목
plt.ylabel("와이 축", fontdict=font3) # y축 제목
plt.show()  # 그래프 화면 표시
plt.close() # 그래프 종료
```

## #04. 스타일 지정
linestyle

### 1. 선 스타일
| 선 스타일 문자열 | 의미 |
|-|-|
| `-` | 실선(solid)|
| `--` |  대시선(dashed)|
| `:` |  점선(dotted)|
| `-.` |  대시-점선(dash-dit)|

```python
plt.figure()    # 그래프 시작을 선언
plt.title("선 스타일 지정")    # 그래프 제목 표시
plt.plot([10,20,30,40,50],[1,4,10,7,15], linestyle='--')   # 그래프 그리기
plt.xlabel("엑스 축") # x축 제목
plt.ylabel("와이 축") # y축 제목
plt.grid()  # 그리드(바탕 줄) 표시
plt.show()  # 그래프 화면 표시
plt.close() # 그래프 종료
```

### 2. 마커

```python
plt.figure()    # 그래프 시작을 선언
plt.title("데이터에 마커 지정")    # 그래프 제목 표시
plt.plot([10,20,30,40,50],[1,4,10,7,15], linestyle='--', marker="^", color = "red")   # 그래프 그리기
plt.grid()  # 그리드(바탕 줄) 표시
plt.show()  # 그래프 화면 표시
plt.close() # 그래프 종료
```

### 3. 기타 스타일
linestyle, linewidth,
marker, marker, markersize, markerfacecolor, markeredgecolor, markeredgewidth

| 스타일 문자열 | 약자 | 의미 |
|-|-|-|
| `color` | `c`  | 선 색깔 |
| `linestyle` | `ls` | 선 스타일 |
| `linewidth` | `lw` | 선 굵기 |
| `marker` |   | 마커 종류 |
| `markersize` | `ms`  | 마커 크기 |
| `markerfacecolor` | `mfc`   |마커 내부 색깔 |
| `markeredgecolor` | `mec`   |마커 선 색깔 |
| `markeredgewidth` | `mew`   |마커 선 굵기 |

```python
plt.figure()    # 그래프 생성 선언
plt.title("스타일 지정2 디테일")    # 제목 선언
# 그래프 그리기
plt.plot([10,20,30,40,50],[1,4,10,7,15],
        ls = '--', lw = 3,
        marker = "*", ms = 20, mfc = 'yellow',
        mec = 'black', mew = 2)
plt.xlabel("X 축")  # x축 이름
plt.ylabel("Y 축")  # y축 이름
plt.grid()  # 바탕줄 표시
plt.show()      # 그래프 시각화
plt.close()     # 그래프 종료
```

## #05. 축 설정
xlim, ylim

x축의 범위(xlim)와 y축의 범위(ylim)을 설정

```python
plt.figure()    # 그래프 생성 선언
plt.title("x축, y축 범위 설정") # 그래프 제목 선언
# plot 그래프 생성
plt.plot([10,20,30,40,50],[1,4,10,7,15])
plt.xlim(0, 60)     # x축의 범위
plt.ylim(-10, 30)     # y축의 범위
plt.grid()  # 바탕선 설정
plt.show()  # 그래프 시각화
plt.close() # 그래프 종료
```

2. 각 축의 표시 내용 설정
xticks, yticks

```python
# y축 범위와 크기 지정
y_values = list(range(70, 101, 5))
y_values
```

```python
# y축 값 지정
y_ticks = []

for i in y_values:
    y_ticks.append("%d점"%i)
y_ticks
```

```python
# 그래프에 적용
plt.figure()    # 그래프 생성 설정
plt.title("학년별 평균 점수 변화")  # 그래프 제목
plt.plot([1,2,3,4],[98,82,73,92])

# x축에 표시될 내용 설정 : 좌표 값 리스트와 표시 내용 리스트
plt.xticks([1,2,3,4], ['1학년','2학년','3학년','4학년'])

plt.grid()
plt.show()
plt.close()
```

```python
# 그래프에 적용
plt.figure()    # 그래프 생성 설정
plt.title("학년별 평균 점수 변화")  # 그래프 제목
plt.plot([1,2,3,4],[98,82,73,92])

# x축에 표시될 내용 설정 : 좌표 값 리스트와 표시 내용 리스트
plt.xticks([1,2,3,4], ['1학년','2학년','3학년','4학년'])
# y축에 표시될 내용 설정 : 좌표 값 리스트와 표시 내용 리스트
plt.yticks(y_values, y_ticks)
plt.grid()
plt.show()
plt.close()
```
