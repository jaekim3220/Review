# 문제1

x축의 간격을 10일로 설정하여 날짜별 종가와 시가 변화를 구현

## 1. 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/D01/bitcoin.xlsx")
df
```

## 2. x 좌표를 위한 시각화.

날짜를 10일 간격이라고 조건이 있기 때문에 x 좌표를 위한 list를 생성

다만 10개 단위로 x에 표현하는 것은 xtick에서 구현

그렇지 않으면 처음부터 10 단위로 좌표를 축력해 값이 변경됨

### x좌표를 위한 리스트 생성

```python
x = list(range(0, len(df['날짜'])))
print(x)
```

### xticks 함수에서 사용할 x좌표 생성

```python
xticks = list(range(0, len(df['날짜']), 50))
print(xticks)
```

### xticks 함수에서 출력할 날짜값 추출

```python
date = []

for i in xticks:
    date.append(df['날짜'][i].strftime("%y/%m/%d"))

date
```

## 3. Line Plot 그리기

```python
plt.figure()

plt.plot(x, df['종가'])
plt.plot(x, df['시가'])
plt.xticks(xticks, date)

plt.grid()
plt.show()
plt.close()
```

# 문제 3

## 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/D01/city_people.xlsx")
df
```

## 데이터 전처리

2015, 2010년 데이터 추출

```python
df2 = df.query("연도 == 2015 or 연도 == 2010")
df2
```

## 피벗 테이블

```python
df3 = pivot_table(df2, index='도시', columns='연도', values='인구')
df3
```

```python
x = np.array(range(0, len(df3.index)))
x
```

## 시각화

```python
# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (10, 5)

plt.figure()

# 각 막대의 색상을 단색으로 지정
plt.bar(x-0.2, df3[2010], color='#0066ff', label="2010", width=0.4)
plt.bar(x+0.2, df3[2015], color='#ff6600', label="2015", width=0.4)

plt.grid()
plt.legend()
plt.xticks(x, df3.index)
plt.ylim(2000000, 10000000)

plt.show()
plt.close()
```

# 문제 5

## 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc.xlsx")
df
```

```python
plt.rcParams["figure.figsize"] = (15, 7)

plt.figure()
plt.scatter(df['발생건수'], df['부상자수'])
plt.xlabel("발생건수")
plt.ylabel("부상자수")
plt.grid()
plt.show()
plt.close()
```

# 추가내용 - numpy를 활용한 추세선 구하기 (회귀분석)

```python
z = np.polyfit(df['발생건수'], df['부상자수'], 1)
print("기울기(가중치) =", z[0])
print("절편(편향) = ", z[1])
```

```python
# 발생건수에 대한 부상자수 예측
print("f(x) = %f * x + %f" % (z[0], z[1]))
```

```python
f = np.poly1d(z)
f
```

```python
f(1)
```

```python
f(df['발생건수'])
```

```python
plt.rcParams["figure.figsize"] = (15, 7)

plt.figure()
plt.scatter(df['발생건수'], df['부상자수'])
plt.plot(df['발생건수'], f(df['발생건수']), color='#ff0000')
plt.xlabel("발생건수")
plt.ylabel("부상자수")
plt.grid()
plt.show()
plt.close()
```
