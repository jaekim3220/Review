# 서브플롯

하나의 그래픽 영역을 나눠 두 개 이상의 시각화 결과를 하나의 화면에서 표현 가능

패키지 설치

```python
import numpy as np
from pandas import read_excel
from matplotlib import pyplot as plt
```

그래프 전역 환경 설정

```python
# 폰트
plt.rcParams['font.family'] =  "Malgun Gothic"  # 윈도우 전용
# 글자 크기
plt.rcParams['font.size'] = 10
# 그래프 크기
plt.rcParams['figure.figsize'] = (12,8)
# 유니코드에서 음수 부호 설정
plt.rcParams['axes.unicode_minus'] = False
```

데이터 준비

```python
df = read_excel("http://data.hossam.kr/D01/traffic_acc_year.xlsx")
df
```

```python
# 데이터 확인 후 index를 설정(년도)
df = read_excel("http://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col= "년도")
df
```

서브플롯 영역 분리

```python
# 2행 3열 서브플롯
# fig 객체 : 그래픽 처리 기능
# ax 객체 : 분할된 각 그래프 영역
fig, ax = plt.subplots(2,3)

# print(ax)

# 전체 제목 생성
fig.suptitle("서브플롯 영역", fontsize=30, color='green')

# 각 그래프의 가로(wspace), 세로(hspace) 간격 지정
fig.subplots_adjust(wspace=0.2, hspace=0.2)

plt.show()
plt.close()
```

## 서브플롯에 그래프 그리기

```python
df
```

```python
# 2행 2열 영역 생성
# figsize 파라미터는 그래픽의 크기
# plt.rcParams["figure.figsize"] 설정보다 우선
fig, ax = plt.subplots(2,2, figsize = (20,12))

# 전체 제목
fig.suptitle("2행 2열의 서브플롯 복습", fontsize=28, color='red')

# 데이터프레임 기반으로 각 영역에 다양한 그래프 표시
df.plot(ax=ax[0][0])
df.plot.bar(ax=ax[0][1])
df['발생건수'].plot.pie(ax=ax[1][0])
df.plot.scatter(x='발생건수',y='부상자수', ax=ax[1][1])

plt.show()
plt.close()
```

## 서브플롯 영역에 옵션 지정

```python
df
```

```python
# 길다
#--------------------------------------------------------------
# 2행 2열의 영역을 생성
#--------------------------------------------------------------
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(30, 15))
fig.suptitle("서브플롯 영역 옵션 지정", fontsize=28, color='blue')

#--------------------------------------------------------------
# 첫 번째 = 선그래프
#--------------------------------------------------------------
df.plot(ax=ax1, color=['#66ff00', '#ff6600', '#0000ff'])

# 제목의 텍스트, 크기, 색상 지정
ax1.title.set_text("교통사고 발생 건수 변화")
ax1.title.set_fontsize(24)
ax1.title.set_color("#ff0000")

# 라벨 지정
ax1.set(xlabel="년도", ylabel="교통사고 수")


# plt을 직접 사용할 경우 xticks(x축좌표, 표시할_텍스트) 형식으로 지정
# 서브플롯은 x축의 좌표와 표시할 텍스트를 나누어서 처리
start = min(df.index)
end = max(df.index)
x = list(range(start, end+1))
# 좌표 설정(x 축에 넣을 값)
ax1.set_xticks(x)
# 지정된 x축 좌표에 표시될 텍스트 리스트를 지정
ax1.set_xticklabels(df.index, fontsize=12, color='#ff0000')
# x축, y축의 범위
ax1.set_xlim([start-0.5, end+0.5])
ax1.set_ylim([0, 400000])

#--------------------------------------------------------------
# 두 번째 영역 그래프 표시 -> 막대 그래프
#--------------------------------------------------------------

df.plot.bar(ax=ax2, rot=0, color=['#ff6633', '#000000', '#0066ff'])
ax2.grid()

# 제목 텍스트, 크기, 색상
ax2.title.set_text("2005~2018년 교통사고 현황")
ax2.title.set_fontsize(24)
ax2.title.set_color("#0000ff")

# x축의 좌표만 지정 (막대그래프의 경우 x축 좌표는 무조건 0부터 시작)
ax2.set_xticks(list(range(0, len(df.index))))
# 지정된 x축 좌표에 표시될 텍스트 리스트를 지정(x 축에 넣을 값)
ax2.set_xticklabels(df.index, fontsize=12, color='#0000ff')

# x축과 y축의 라벨을 지정 (불필요한 부분은 생략 가능)
ax2.set(xlabel='년도', ylabel='교통사고')

#--------------------------------------------------------------
# 세 번째 영역 그래프 표시 -> 파이 그래프
#--------------------------------------------------------------


df['발생건수'].plot.pie(ax=ax3, labels = df.index, autopct="%0.1f%%",
                    textprops={'color':"#ff6600", 'fontsize': 12})

# 그래프 제목
ax3.title.set_text("년도별 교토사고 발생비율")

# 파이 그래프는 격자 표시 안 됨

# 여기서 y축 라벨은 불필요하므로 제거
ax3.set(ylabel=None)

# 범주 표시
# title로 범주 제목 설정, bbox_to_anchor로 범주 위치 설정(loc 함수도 가능)
ax3.legend(labels=df.index, title='범주', bbox_to_anchor=(1.4, 0.9))

#--------------------------------------------------------------
# 네 번째 영역 그래프 표시 -> 산점도 그래프
#--------------------------------------------------------------

# 산점도 그래프
df.plot.scatter(ax=ax4, x='발생건수', y = '부상자수',
                color="#ff6600", marker ="*", label='교통사고')

# 배경 격자
ax4.grid()

# x축, y축 라벨
ax4.set(xlabel="발생 건수", ylabel='부상자수')

#--------------------------------------------------------------
# 그래프 출력하기
#--------------------------------------------------------------

plt.show()
plt.close()
```

## 두 개의 y축을 갖는 그래프

샘플 데이터

```python
x = list(range(0,10))
# x
y1 = list(range(0,10,1))
y2 = list(np.arange(100, 1001, 100))
# y2 = list(np.arange(0, 1000, 100))는 완전 겹쳐서 확인이 안 됨
print(y1,y2)
```

기본 그래프

```python
plt.rcParams['figure.figsize'] = (7,4)

# 기울기 차이가 발생
plt.figure()
plt.plot(x,y1, label='y1', color="#ff00ff")
plt.plot(x,y2, label='y2', color='#0000ff')
plt.legend()
plt.show()
plt.close()
```

서브플롯으로 2개의 y축을 갖는 그래프 구현, 이전과 비교

```python
# subplot 아님
fig, ax1 = plt.subplots(1,1, figsize=(7,4))

# ax1에 겹쳐지는 쌍둥이 서브플롯을 생성
ax2 = ax1.twinx()

# ax1에 그래프 그리기
ax1.plot(x, y1, color="#ff00ff")
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='#ff00ff')
ax1.grid()
ax1.set_ylim([0, 10])

# ax2에 그래프 그리기
ax2.plot(x, y2, color="#0000ff")
ax2.set_ylabel('Y2 data', color='#0000ff')
ax2.set_ylim([0, 1000])
```

기존 데이터 사용

```python
fig, ax1 = plt.subplots(1,1, figsize = (14,8))
ax2 = ax1.twinx()

# ax1에 그래프 그리기
df['발생건수'].plot(ax=ax1, color="#ff0000")
ax1.set_xlabel("년도")
ax1.set_ylabel("발생건수", color='#ff0000')
ax1.grid()

# x축 좌표 텍스트 처리
start = min(df.index)
end = max(df.index)
ax1.set_xticks(list(range(start,end+1)))    #x축 개수 설정
ax1.set_xticklabels(df.index, fontsize=12, color='#000000') #x축 값 설정

# ax2 그래프 그리기
df['부상자수'].plot(ax=ax2, color='#0000ff')
ax2.set_ylabel('부상자수', color='#0000ff') #x label은 ax1과 동일

plt.show()
plt.close()
```
