# 데이터프레임 시각화

데이터프레임이 내부적으로 `pyplot` 객체를 초기화해 `plt.figure()` 호출할 필요 없음

패키지 설치

```python

from pandas import read_excel
from matplotlib import pyplot as plt
```

데이터 준비

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx")
df
```

```python
# 추출한 데이터 확인 후 index를 설정 결정
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col='년도')
df 
```

그래프 전역 설정

```python
# 폰트
plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 전용
#plt.rcParams['font.family'] = 'AppleGothic'   # 맥 전용
# 글자 크기
plt.rcParams["font.size"] = 10
# 그래프 크기
plt.rcParams["figure.figsize"] = (10, 5)
# 음수 설정
plt.rcParams['axes.unicode_minus'] = False
```

## Line Plot

특정 컬럼에 대한 시각화 - [ ]

```python
# 그래프 생성 전 데이터 확인
df
```

전체 컬럼에 대한 시각화

```python
df.plot()   #plot(선) 그래프 생성
plt.title("교통사고 발생건수")  #그래프 제목
plt.grid()  #바탕 선 생성
plt.legend()    #라벨 생성
plt.show()  #그래프 시각화
plt.close() #그래프 종료
```

```python
# 발생건수 데이터 추출
df['발생건수'].plot()   #plot(선) 그래프 생성
plt.title("교통사고 발생건수")  #그래프 제목
plt.grid()  #바탕 선 생성
plt.legend()    #라벨 생성
plt.show()  #그래프 시각화
plt.close() #그래프 종료
```

위의 그래프와 label의 유무를 비교

```python
df['발생건수'].plot(label='교통사고')
plt.title("교통사고 발생건수")
plt.grid()
plt.legend()
plt.show()
plt.close()
```

2개 이상 컬럼에 대한 시각화 - filter

```python
df.filter(['사망자수', '부상자수']).plot()
plt.title("교통사고 사망자수, 부상자수")
plt.grid()
plt.legend()
plt.show()
plt.close()
```

## Bar Plot

전체 컬럼 데이터

```python
df.plot.bar(rot=45)
plt.grid()
plt.legend()
plt.title("년도별 교통사고 변화")
plt.xlabel("연도")
plt.ylabel("숫자")
plt.show()
plt.close()
```

특정 컬럼 데이터 시각화

```python
df['발생건수'].plot.bar()
plt.grid()
plt.title("년도별 교통사고 변화")
plt.xlabel("연도")
plt.ylabel("숫자")
plt.ylim(200000, 240000)    #y축 값 설정
plt.show()
plt.close()
```

rot를 사용해 x 축 값 회전

```python
df['발생건수'].plot.bar(rot=45)
plt.title("년도별 교통사고 변화")
plt.ylim(200000, 240000)    #y축 값 설정
plt.xlabel("연도")
plt.ylabel("숫자")
plt.grid()
plt.show()
plt.close()
```

2개 이상 컬럼에 대한 시각화 - filter

```python
df.filter(['발생건수','부상자수']).plot.bar(rot=45) #여러 컬럼을 삽입
plt.title("년도별 교통사고 변화")
plt.xlabel("연도")
plt.ylabel("숫자")
plt.grid()
plt.legend()
plt.ylim(150000, 370000)    #y축 값 설정
plt.show()
plt.close()
```

## 산점도 그래프

`두 집단`이 상관 관계가 있는 경우 사용

```python
df.plot.scatter(x='발생건수', y='부상자수', color="#0066ff")
plt.title("발생 건수 별 부상자 수 집계")
plt.grid()
plt.show()
plt.close()
```

## 파이 그래프

dataframe의 index가 파이 그래프의 각 조각이 된다

```python
# 하나의 컬럼을 지정해서 데이터 시각화
df['사망자수'].plot.pie(autopct='%0.1f%%')
plt.title("연도별 사망자 수 비율")
plt.show()
plt.close()
```
