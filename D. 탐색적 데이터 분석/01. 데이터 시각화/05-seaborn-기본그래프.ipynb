# seaborn - 기본 그래프

주요 파라미터

| 파라미터 | 설명 |
|---|---|
| `data`ㅤ| 데이터 구조 입력. 일반적으로 데이터프레임 객체 |
| `x`, `y`ㅤ| x, y축의 값을 지정하는 변수<br/>`data`를 설정하지 않은 경우 리스트 등의 연속형 변수<br/>`data`를 설정한 경우 데이터프레임의 컬럼이름 |
| `hue` | 데이터를 구분하기 위한 그룹화 변수(범주형 변수)<br/>`data`를 설정한 경우 데이터프레임의 컬럼이름 |
| `size`ㅤ| 너비가 다른 선을 생성하는 그룹화 변수 (마커의 경우, 마커의 크기가 다르게 나타남) |
| `style`ㅤ| 다른 스타일의 마커 또는 선을 생성하는 그룹화 변수 |
| `palette`| 색상 체계 선택 |

패키지 준비

```python
from matplotlib import pyplot as plt
from pandas import read_excel
import seaborn as sb
```

그래프 전역 설정

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.unicode_minus"] = False
```

## 기본 그래프

pie 그래프는 seaborn에서  제공 안됨

1. 선그래프 plot

```python
# 그래프 그리기
sb.lineplot(x=[-20,-10,0,10,20],y=[1,3,5,7,9], label='data1')
sb.lineplot(x=[-20,-10,0,10,20],y=[1,2,3,4,5], label='data2')

# 그 외 옵션은 matplotlib와 동일
plt.title("한글 제목")
plt.xlabel("엑스 축")
plt.ylabel("와이 축")
plt.grid()
plt.legend()
plt.show()
plt.close()
```

2. 세로 막대 그래프

```python
# 데이터 설정
x = ['서울', '부산', '제주']
y = [20,30,10]

sb.barplot(x=x, y=y, label=['서울','부산','제주'])

plt.title("Bar Chart")
plt.xlabel("엑스 축")
plt.ylabel("와이 축")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

3. 가로 막대 그래프

팔레트 사용
>deep, muted, pastel, bright, dark, colorblind, Spectral

```python
# 데이터 설정(barh와는 다르게 x,y 위치 바꿀 필요 없음)

x = ['서울', '부산', '제주']
y = [20,30,10]

sb.barplot(x=y, y=x, label=['서울','부산','제주'],
           palette='muted')

plt.title("Bar Chart2")
plt.xlabel("엑스 축")
plt.ylabel("와이 축")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

```python
# palette 비교
x = ['서울', '부산', '제주']
y = [20,30,10]

sb.barplot(x=y, y=x, label=['서울','부산','제주'],
           palette='colorblind')

plt.title("Bar Chart2")
plt.xlabel("엑스 축")
plt.ylabel("와이 축")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

4. 산점도 그래프

```python
# 온도와 아이스크림 판매 수량
tmp = [23, 25, 26, 27, 28, 29, 30, 31, 33]
qty = [21, 23, 25, 28, 33, 35, 36, 32, 39]

sb.scatterplot(x=tmp, y=qty, label='판매량')

plt.title("Scatter Chart")
plt.xlabel("온도")
plt.ylabel("판매량")
plt.legend()
plt.grid()
plt.show()
plt.close()
```

## 데이터 프레임을 통한 그래프

데이터 준비

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx")
df
```

```python
# 데이터 확인 후 index 결정
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col="년도")
df
```

2. 선 그래프

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)

sb.lineplot(data=df, x=df.index, y='사망자수', label='사망자수')

plt.title('교통사고 사망자수')   # 그래프 제목
plt.xlabel("년도")      # x축 제목
plt.ylabel("교통사고")   # y축 제목
plt.legend()
plt.grid()
plt.show()               # 그래프 화면 표시
plt.close()              # 그래프 그리기 종료
```

3. 막대 그래프

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.barplot(data=df, x=df.index, y='사망자수', label='사망자수')

plt.title('교통사고 사망자수')  # 그래프 제목
plt.xlabel("년도")  # x축 제목
plt.ylabel("교통사고")  # y축 제목
plt.legend()    # 범주표시
plt.grid()  # 격자표시
plt.show()  # 그래프 화면 표시
plt.close()   
```

4. 산점도 그래프

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.scatterplot(data=df, x='발생건수', y='사망자수', label='사망자수')

plt.title('교통사고 발생현황')  # 그래프 제목
plt.xlabel("발생건수")  # x축 제목
plt.ylabel("사망자수")  # y축 제목
plt.legend()    # 범주표시
plt.grid()  # 격자표시
plt.show()  # 그래프 화면 표시
plt.close() 
```

5. 추세선을 포함한 산점도 그래프

추세선 주변 구간은 95% 신뢰 구간

fit_reg=False를 추가 시 추세선이 제외, scatterplot 메서드와 동일 결과를 제공(기본값=True)

line_kws 파라미터는 선의 색상, scatter_kws는 점의 색상

```python
# 추세선을 함께 보여주는 산점도 그래프 생성
sb.regplot(data=df, x='발생건수', y='사망자수', label='사망자수',
           line_kws={'color': '#301ce6'}, scatter_kws={'color': '#ed1a36'})

plt.title("교통사고 발생 현황")
plt.xlabel("발생건수")
plt.ylabel("사망자수")
plt.legend()    # 범주표시
plt.grid()  # 격자표시
plt.show()  # 그래프 화면 표시
plt.close() 
```

5. 서브플롯에 기반한 선 그래프 - 2개의 y 값

> 서브 플롯 사용하지 않을 경우 sb.lineplot() 함수에서 ax 파라미터 제거

```python
fig, ax1 = plt.subplots(1,1)
ax2 = ax1.twinx()

# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.lineplot(data=df, x=df.index, y='발생건수', color='#ff6600', ax=ax1)
sb.lineplot(data=df, x=df.index, y='부상자수', color='#0066ff', ax=ax2)

plt.title("교통사고 발생현황")
plt.xlabel("년도")
plt.ylabel("교통사고")
plt.grid()
plt.show()
plt.close()
```
